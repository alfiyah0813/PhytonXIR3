```python
from abc import ABC, abstractmethod

class BarangElektronik(ABC):
    def __init__(self, nama, harga_dasar):
        self.nama = nama
        self.__harga_dasar = harga_dasar
        self.__stok = 0  # private attribute

    # Getter untuk stok
    def get_stok(self):
        return self.__stok

    # Method untuk menambah stok dengan validasi
    def tambah_stok(self, jumlah):
        if jumlah < 0:
            print(f"Gagal update stok {self.nama}! Stok tidak boleh negatif ({jumlah}).")
        else:
            self.__stok += jumlah
            print(f"Berhasil menambahkan stok {self.nama}: {self.__stok} unit.")

    # Getter untuk harga dasar
    def get_harga_dasar(self):
        return self.__harga_dasar

    # Abstract method wajib di-override anak
    @abstractmethod
    def tampilkan_detail(self):
        pass

    @abstractmethod
    def hitung_harga_total(self, jumlah):
        pass
```

1. BarangElektronik adalah abstract class yang menjadi kerangka semua barang elektronik.
2. Stok dan harga dasar dijaga privat (__stok, __harga_dasar) => Encapsulation.
3. tambah_stok() menolak stok negatif.
4. Method abstract tampilkan_detail() & hitung_harga_total() wajib diimplementasikan anak class.



```python
class Laptop(BarangElektronik):
    def __init__(self, nama, harga_dasar, processor):
        super().__init__(nama, harga_dasar)
        self.processor = processor

    def tampilkan_detail(self):
        print(f"[LAPTOP] {self.nama} | Proc: {self.processor}")
        print(f"Harga Dasar: Rp {self.get_harga_dasar():,} | Pajak(10%): Rp {int(self.get_harga_dasar()*0.1):,}")

    def hitung_harga_total(self, jumlah):
        pajak = 0.1 * self.get_harga_dasar()
        return (self.get_harga_dasar() + pajak) * jumlah
```

1. Class Laptop mewarisi BarangElektronik
2. Menambahkan atribut processor.
3. Pajak 10% diterapkan dalam hitung_harga_total().
4. tampilkan_detail() dioverride untuk menampilkan info laptop.

```python
class Smartphone(BarangElektronik):
    def __init__(self, nama, harga_dasar, kamera):
        super().__init__(nama, harga_dasar)
        self.kamera = kamera

    def tampilkan_detail(self):
        print(f"[SMARTPHONE] {self.nama} | Cam: {self.kamera}")
        print(f"Harga Dasar: Rp {self.get_harga_dasar():,} | Pajak(5%): Rp {int(self.get_harga_dasar()*0.05):,}")

    def hitung_harga_total(self, jumlah):
        pajak = 0.05 * self.get_harga_dasar()
        return (self.get_harga_dasar() + pajak) * jumlah
```

1. Class Smartphone juga merupakan turunan dari BarangElektronik
2. Atribut tambahan: kamera
3. Pajak 5% diterapkan di method hitung_harga_total().

```python
print("--- SETUP DATA ---")

# Buat produk
laptop = Laptop("ROG Zephyrus", 20000000, "Ryzen 9")
smartphone = Smartphone("iPhone 13", 15000000, "12MP")

# Tambah stok
laptop.tambah_stok(10)
smartphone.tambah_stok(-5)  # dicoba input negatif
smartphone.tambah_stok(20)
```

1. Pada code ini admin membuat data produk dan mencoba menambahkan stok
2. Jika stok negatif, program menolak dan menampilkan peringatan.
3. Stok berhasil ditambahkan jika valid.

```python
def proses_transaksi(daftar_barang):
    print("\n--- STRUK TRANSAKSI ---")
    total = 0

    for i, (barang, jumlah) in enumerate(daftar_barang, start=1):
        barang.tampilkan_detail()
        subtotal = barang.hitung_harga_total(jumlah)
        print(f" Beli: {jumlah} unit | Subtotal: Rp {int(subtotal):,}")
        print()
        total += subtotal

    print("-" * 40)
    print(f"TOTAL TAGIHAN: Rp {int(total):,}")
```

1. Fungsi dari kode ini adalah menerima list berisi campuran objek Laptop dan juga Smartphone
2. Total belanja dihitung otomatis menggunakan methode polymorphism.
3. Code ini menampilkan subtotal & total tagihan.

```python
keranjang = [
    (laptop, 2),       # User beli 2 laptop
    (smartphone, 1)    # User beli 1 smartphone
]

proses_transaksi(keranjang)
```

```python
# 1. Import dan Abstract Class
from abc import ABC, abstractmethod

class BarangElektronik(ABC):
    def __init__(self, nama, harga_dasar):
        self.nama = nama
        self.__harga_dasar = harga_dasar
        self.__stok = 0

    def get_stok(self):
        return self.__stok

    def tambah_stok(self, jumlah):
        if jumlah < 0:
            print(f"Gagal update stok {self.nama}! Stok tidak boleh negatif ({jumlah}).")
        else:
            self.__stok += jumlah
            print(f"Berhasil menambahkan stok {self.nama}: {self.__stok} unit.")

    def get_harga_dasar(self):
        return self.__harga_dasar

    @abstractmethod
    def tampilkan_detail(self):
        pass

    @abstractmethod
    def hitung_harga_total(self, jumlah):
        pass


# 2. Class Anak
class Laptop(BarangElektronik):
    def __init__(self, nama, harga_dasar, processor):
        super().__init__(nama, harga_dasar)
        self.processor = processor

    def tampilkan_detail(self):
        print(f"[LAPTOP] {self.nama} | Proc: {self.processor}")
        print(f"Harga Dasar: Rp {self.get_harga_dasar():,} | Pajak(10%): Rp {int(self.get_harga_dasar()*0.1):,}")

    def hitung_harga_total(self, jumlah):
        pajak = 0.1 * self.get_harga_dasar()
        return (self.get_harga_dasar() + pajak) * jumlah

class Smartphone(BarangElektronik):
    def __init__(self, nama, harga_dasar, kamera):
        super().__init__(nama, harga_dasar)
        self.kamera = kamera

    def tampilkan_detail(self):
        print(f"[SMARTPHONE] {self.nama} | Cam: {self.kamera}")
        print(f"Harga Dasar: Rp {self.get_harga_dasar():,} | Pajak(5%): Rp {int(self.get_harga_dasar()*0.05):,}")

    def hitung_harga_total(self, jumlah):
        pajak = 0.05 * self.get_harga_dasar()
        return (self.get_harga_dasar() + pajak) * jumlah


# 3. Setup Data & Tambah Stok
print("--- SETUP DATA ---")
laptop = Laptop("ROG Zephyrus", 20000000, "Ryzen 9")
smartphone = Smartphone("iPhone 13", 15000000, "12MP")

laptop.tambah_stok(10)
smartphone.tambah_stok(-5)
smartphone.tambah_stok(20)


# 4. Fungsi Proses Transaksi
def proses_transaksi(daftar_barang):
    print("\n--- STRUK TRANSAKSI ---")
    total = 0
    for i, (barang, jumlah) in enumerate(daftar_barang, start=1):
        barang.tampilkan_detail()
        subtotal = barang.hitung_harga_total(jumlah)
        print(f" Beli: {jumlah} unit | Subtotal: Rp {int(subtotal):,}\n")
        total += subtotal
    print("-" * 40)
    print(f"TOTAL TAGIHAN: Rp {int(total):,}")


# 5. Simulasi Pembelian
keranjang = [
    (laptop, 2),
    (smartphone, 1)
]

proses_transaksi(keranjang)


# 6. Produk Tambahan
laptop2 = Laptop("MacBook Pro", 700000000000000, "M1 Pro")
smartphone2 = Smartphone("Samsung Galaxy S23 PRO", 350000000000000, "108MP")

laptop2.tambah_stok(5)
smartphone2.tambah_stok(8)

keranjang2 = [
    (laptop2, 1),
    (smartphone2, 2)
]

proses_transaksi(keranjang2)
```
