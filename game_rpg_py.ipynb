```python
class Hero:
    # Constructor: Dijalankan saat Hero baru dibuat
    def __init__(self, name, hp, attack_power):
        self.name = name          # Nama Hero
        self.hp = hp              # Nyawa (Health Point)
        self.attack_power = attack_power  # Kekuatan Serangan

    # Method untuk menampilkan info hero
    def info(self):
        print(f"Hero: {self.name} | HP: {self.hp} | Power: {self.attack_power}")

# -- Main Program --
# Membuat Object (Instansiasi)
hero1 = Hero("Layla", 500, 15)
hero2 = Hero("Zilong", 120, 20)

# Memanggil Method
hero1.info()
hero2.info()
```

**Jawaban : **
hero1.hp adalah atribut dari objek hero1. Meskipun awalnya hero1.hp bernilai 100 saat objek dibuat, kita bisa mengubahnya kapan saja. Setelah diubah, nilai HP hero1 menjadi 500, sehingga print(hero1.hp) menampilkan 500.

```python
class Hero:
    def __init__(self, name, hp, attack_power):
        self.name = name
        self.hp = hp
        self.attack_power = attack_power

    def info(self):
        print(f"Hero: {self.name} | HP: {self.hp} | Power: {self.attack_power}")

    # Method menyerang: Objek ini (self) menyerang objek lain (lawan)
    def serang(self, lawan):
        print(f"{self.name} menyerang {lawan.name}!")
        lawan.diserang(self.attack_power)

    # Method diserang: Menerima damage
    def diserang(self, damage):
        self.hp -= damage
        print(f"{self.name} terkena damage {damage}. Sisa HP: {self.hp}")


# -- Main Program --
hero1 = Hero("Layla", 100, 15)
hero2 = Hero("Zilong", 120, 20)

print("\n--- Pertarungan Dimulai ---")
hero1.serang(hero2)  # Layla menyerang Zilong
hero2.serang(hero1)  # Zilong membalas
```

**Jawaban :**

Lawan adalah objek Hero, sehingga kita bisa mengakses seluruh atribut dan metode lawan. Contohnya, kita bisa memanggil lawan.diserang(self.attack_power) untuk mengurangi HP lawan. Jika lawan hanya string (misalnya "Zilong"), kita tidak bisa mengubah HP lawan atau memanggil method diserang, karena string tidak memiliki atribut hp atau method untuk diserang(). Dengan menerima objek, metode serang bisa berinteraksi sepenuhnya dengan hero lawan, termasuk HP, serangan balik, atau skill lain di masa depan. *Maksudnya dengan menerima objek memungkinkan untuk interaksi realistis antara hero, bukan sekadar menampilkan nama.*

```python
class Mage(Hero):
    def __init__(self, name, hp, attack_power, mana):
        super().__init__(name, hp, attack_power)
        self.mana = mana

    def info(self):
        print(f"{self.name} [Mage] | HP: {self.hp} | Mana: {self.mana}")

    def skill_fireball(self, lawan):
        if self.mana >= 20:
            print(f"{self.name} menggunakan Fireball ke {lawan.name}!")
            self.mana -= 20
            lawan.diserang(self.attack_power * 2)
        else:
            print(f"{self.name} gagal skill! Mana tidak cukup.")


# Main Program
eudora = Mage("Eudora", 80, 30, 100)
eudora.info()
```

Atribut name, hp, dan attack_power milik class Hero tidak dibuat, karena constructor parent tidak dipanggil. Jadi meskipun kita memberikan "Eudora" saat membuat objek, nilai itu tidak tersimpan di objek. Maka dari itu peran dari super() adalah untuk memanggil constructor dari class induk, sehingga objek anak (Mage) mewarisi semua atribut dan method dari parent (Hero). Tanpa super(), objek anak hanya punya atribut yang didefinisikan sendiri (misal mana) dan atribut parent tidak tersedia.

```python
class Hero:
    def __init__(self, nama, hp_awal):
        self.nama = nama
        # Enkapsulasi: HP bersifat Private (hanya bisa diakses di dalam class ini)
        self.__hp = hp_awal

    # GETTER: Cara resmi melihat HP
    def get_hp(self):
        return self.__hp

    # SETTER: Cara resmi mengubah HP (dengan validasi)
    def set_hp(self, nilai_baru):
        if nilai_baru < 0:
            self.__hp = 0  # HP tidak boleh negatif
        elif nilai_baru > 1000:
            print("Cheat terdeteksi! HP dimaksimalkan ke 1000 saja.")
            self.__hp = 1000
        else:
            self.__hp = nilai_baru

    # Method untuk menerima damage
    def diserang(self, damage):
        # Kita pakai setter/getter bahkan di dalam class sendiri agar aman
        sisa_hp = self.get_hp() - damage
        self.set_hp(sisa_hp)
        print(f"{self.nama} terkena damage {damage}. Sisa HP: {self.get_hp()}")


# -- Uji Coba --
hero1 = Hero("Layla", 100)

# hero1.__hp = 9999  # GAGAL (Tidak akan mengubah hp asli)
# print(hero1.__hp)   # ERROR (Tidak bisa dibaca langsung)

hero1.set_hp(-50)      # Coba set negatif
print(hero1.get_hp())  # Output: 0 (Karena dicegat oleh logika Setter)
```

```python
from abc import ABC, abstractmethod

# 1. Interface / Abstract Class
# Ini adalah KONTRAK. Semua turunan wajib punya method di bawah ini.
class GameUnit(ABC):

    @abstractmethod
    def serang(self, target):
        pass

    @abstractmethod
    def info(self):
        pass

# 2. Implementasi pada Class Konkret
class Hero(GameUnit):
    def __init__(self, nama):
        self.nama = nama

    # Kita WAJIB membuat method serang
    def serang(self, target):
        print(f"Hero {self.nama} menebas {target}!")

    def info(self):
        print(f"Saya adalah Hero: {self.nama}")

class Monster(GameUnit):
    def __init__(self, jenis):
        self.jenis = jenis

    # Implementasi serang versi Monster
    def serang(self, target):
        print(f"Monster {self.jenis} menggigit {target}!")

    def info(self):
        print(f"Saya adalah Monster: {self.jenis}")


# -- Uji Coba --
# unit = GameUnit() # ERROR! Abstract class tidak bisa jadi objek.

h = Hero("Alucard")
m = Monster("Serigala")

h.info()
m.info()
```

2. **Jika kita aktifkan:**

**Maka Error yang muncul:**

**Penjelasan:**

Abstract class tidak bisa dijadikan objek langsung karena masih ada method yang belum diimplementasikan.

Tujuan GameUnit: Sebagai template/kontrak bagi semua unit game (Hero, Monster, dll) Menjamin setiap turunan memiliki method serang dan info. Memudahkan programmer bekerja dengan banyak jenis unit secara konsisten tanpa harus tahu detail tiap class

```python
# Parent Class
class Hero:
    def __init__(self, nama):
        self.nama = nama

    def serang(self):
        print("Hero menyerang dengan tangan kosong.")

# Child Class 1
class Mage(Hero):
    def serang(self):
        print(f"{self.nama} (Mage) menembakkan Bola Api! Boom!")

# Child Class 2
class Archer(Hero):
    def serang(self):
        print(f"{self.nama} (Archer) memanah dari jauh! Jleb!")

# Child Class 3
class Fighter(Hero):
    def serang(self):
        print(f"{self.nama} (Fighter) memukul dengan pedang! Slash!")

# Child Class 4 (baru)
class Healer(Hero):
    def serang(self):
        print(f"{self.nama} tidak menyerang, tapi menyembuhkan teman!")

# -- Penerapan Polimorfisme --
pasukan = [
    Mage("Eudora"),
    Archer("Miya"),
    Fighter("Zilong"),
    Mage("Gord"),
    Healer("Angela")  # karakter baru
]

print("--- PERANG DIMULAI ---")
for pahlawan in pasukan:
    pahlawan.serang()
```
